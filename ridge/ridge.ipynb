```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import Ridge
from sklearn.model_selection import train_test_split
```

```python
#let's import our fucking data

data = pd.read_csv("../dataset/1632300362534233.csv")
```

```python
data["Address"].value_counts()
```

```python
print(data["Area"].dtype)
data["Area"]
```

```python
data = data.dropna()
```

```python
for i in data["Area"]:
    print(type(i))
```

```python
data = data[~data["Area"].str.contains(",",na=False)]
```

```python
for i in data["Area"]:
    if "," in i:
        print(i)
```

```python
data["Area"] = data["Area"].astype("Int64")
```

```python
for i in data["Area"]:
    print(type(i))
```

```python
phase0_data = data.drop(columns=["Address","Price"])
```

```python
phase0_train,phase0_test = train_test_split(phase0_data,random_state=0)
```

```python
phase0_train_x = phase0_train.drop(columns=["Price(USD)"])
phase0_train_y = phase0_train["Price(USD)"]

phase0_test_x = phase0_test.drop(columns=["Price(USD)"])
phase0_test_y = phase0_test["Price(USD)"]
```

```python
rid = Ridge(alpha=10000)
rid.fit(phase0_train_x,phase0_train_y)
```

```python
rid.score(phase0_test_x,phase0_test_y)
```
